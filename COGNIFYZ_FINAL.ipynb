# Task 6 :Savings Objective <br>
***Objective:Identify and present main savings objectives***<br>
***Steps:*** <br>
***1.Analyze Saving Objective:*** Examine the column contaning information about participants savings objectives.<br>
***2.List and Describe Objective:*** Create a list and descrbe the main savings objectives mentioned by participants<br>

Analyze the Reasons Columns: Investigate each column containing reasons for investment choices (Reason_Equity, Reason_Mutual, Reason_Bonds, Reason_FD).

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Assuming your dataset is loaded as `dataset`
# Columns containing reasons for investment choices
reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

# Function to count and summarize reasons in each column
def analyze_reasons(column):
    reasons = dataset[column].dropna()  # Drop any NaN values if present
    reason_count = Counter(reasons)
    common_reasons = reason_count.most_common(5)  # Get the top 5 common reasons
    return reason_count, common_reasons

# Plotting function to visualize frequency counts
def plot_reasons(reason_count, title):
    plt.figure(figsize=(10, 5))
    plt.bar(reason_count.keys(), reason_count.values(), color='skyblue')
    plt.xlabel('Reasons')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()

# Display reason counts, plots, and summarize for each column
insights = {}

for col in reason_columns:
    reason_count, common_reasons = analyze_reasons(col)
    
    # Print the frequency count of reasons
    print(f"Frequency count for {col}:")
    print(reason_count)
    print("\n")
    
    # Plot the frequency count
    plot_reasons(reason_count, f"Frequency of Reasons in {col}")
    
    # Store insights for the summary
    insights[col] = common_reasons

# Print summarized insights
print("Summary of Common Themes or Recurring Reasons:")
for col, reasons in insights.items():
    print(f"\n{col} - Common Themes:")
    for reason, count in reasons:
        print(f"  - {reason}: Mentioned {count} times")
```

# Task 10 :Correlation Analysis <br>
***Objective:Explore potential correlations between factors***<br>
***Steps:*** <br>
***1.Select Relevent Columns:*** Identify columns such as age, investment duration,and expected returns for correlation analysis<br>
***2.Use Statistical Methods or Visulizations:*** Employ statistical methods(e.g. correlation coefficients) or visulize potential correlations.

```python
# Print unique values in the 'Expect' column
print("Unique values in 'Expect' column:")
print(dataset['Expect'].unique())

# Convert 'Duration' to numeric
duration_map = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
dataset['Duration_numeric'] = dataset['Duration'].map(duration_map)

# Convert 'Expect' to numeric using the midpoint of the range
def expect_to_numeric(expect_range):
    low, high = map(lambda x: float(x.strip('%')), expect_range.split('-'))
    return (low + high) / 2

dataset['Expect_numeric'] = dataset['Expect'].apply(expect_to_numeric)

# Create correlation matrix
correlation_matrix = dataset[['age', 'Duration_numeric', 'Expect_numeric']].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap: Age, Investment Duration, and Expected Returns')
plt.show()

# Print correlation coefficients
print("\nCorrelation Coefficients:")
print(correlation_matrix)
```

### Correlation Analysis Summary

**Age and Investment Duration:**
- Weak positive relationship (around 0.2 to 0.3).
- Older investors tend to favor longer investment durations.

**Age and Expected Returns:**
- Very minimal influence (close to 0).
- Age doesnâ€™t significantly impact return expectations.

**Investment Duration and Expected Returns:**
- Weak positive relationship (around 0.1 to 0.2).
- Longer horizons align with slightly higher return expectations.

### Key Takeaways:
- **No strong correlations** between age, investment duration, and expected returns.
- Age has only a **minor influence** on investment choices.
- **Longer investment periods** are slightly linked to higher return expectations.
- Other factors like **risk tolerance, goals, and market knowledge** likely play a bigger role.
